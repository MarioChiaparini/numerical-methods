## Runge-Kutta Method (Order 4) - Ordinary Diferential Equation Solution

```julia
using MTH229
using Plots
using Printf
using SymPy
using DifferentialEquations
```

```julia
function rungekutta4(f, y0, t)
    n = length(t)
    y = zeros((n, length(y0)))
    y[1,:] = y0
    for i in 1:n-1
        h = t[i+1] - t[i]
        k1 = f(y[i,:], t[i])
        k2 = f(k1 * h/2 .+ y[i,:], h/2 .+ y[i,:])
        k3 = f(k2 * h/2 .+ y[i,:], h/2 .+ y[i,:])
        k4 = f(k3 * h .+ y[i,:], h .+ y[i,:])
        y[i+1,:] = (h/6) * (k1 .+ 2*k2 .+ 2*k3 .+ k4) .+ y[i,:]
    end
    return y
end

function rgk4(dxdy, tn0, tnf, y0, n)
    tn = LinRange(tn0, tnf, n+1)
    y = zeros(n, length(y0))
    y[1,:] = y0
    for i in 1:n-1
        j = i + 1
        h = tn[j] - tn[i]
        k1 = f(y[i,:], t[i])
        k2 = f(k1 * h/2 .+ y[i,:], h/2 .+ y[i,:])
        k3 = f(k2 * h/2 .+ y[i,:], h/2 .+ y[i,:])
        k4 = f(k3 * h .+ y[i,:], h .+ y[i,:])
        y[j,:] = (h/6) * (k1 .+ 2*k2 .+ 2*k3 .+ k4) .+ y[i,:]
    end
    return y
end 
```
