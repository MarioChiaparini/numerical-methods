## Runge-Kutta Method - Ordinary Diferential Equation Solution

```julia
using MTH229
using Plots
using Printf
using SymPy
```

```julia
function rk4(f::Function, xn::AbstractFloat, x0::Number, y0::Number, n::Integer)
    h = (xn-x0)/n
    @sprintf("solution")
    i=0
    while i < n
        i += 1
        k1 = h * (f(x0,y0))
        k2 = h * (f((x0+h/2), (y0+k1/2)))
        k3 = h * (f((x0+h/2), (y0+k2/2)))
        k4 = h * (f((x0+h), (y0+k3)))
        k =  (k1+2*k2+2*k3+k4)/6
        yn = y0 + k
        @sprintf("--------------------------")
        @sprintf("%.3f\t%.3f\t%.3f", x0,y0,yn)
        y0 = yn
        x0 = x0+h
    @sprintf("\n x=%.3f, y=%3.f",xn,yn)
    end
    return xn,yn 
end
```

```julia
step = 2
xn = 1
y0 = 1
x0 = 0
f(x,y) = (1 .-x').^2 .+ 100 .*(y.-x'.^2).^2
rk4(f,xn,x0,y0,step)
```

```julia
x  = [-2:0.05:2;]
y = [-1:0.05:3;]

Plots.plot(x,y,f,st=:surface,color=cgrad(:jet))
```
