```julia
using MTH229
using Plots
using Printf
using SymPy
```

```julia
b = 0.25
c = 5.0

function f(x,y)
    (-b * y) - (c * sin(y^2))
end

x  = [-2:0.05:2;]
y = [-1:0.05:3;]

Plots.plot(x,y,f,st=:surface,color=cgrad(:jet))
```

```julia
function f(x,y)
    e^(x) + sin(y)*cos(x+y) + x^2
end

x  = [-2:0.05:2;]
y = [-1:0.05:3;]

Plots.plot(x,y,f,st=:surface,color=cgrad(:jet))
```
